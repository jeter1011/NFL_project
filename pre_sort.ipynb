```python
year = 1989  # Variable to track the current draft year
# Iterate over the dataframe rows
for index, row in draft_df.iterrows():
    # Extract the draft year from the "Round" and "Pick" columns
    if row['Pick'] == 1 and row['Round'] == 1:
        year += 1
    draft_df.at[index, 'Year'] = year

draft_df['Year'] = draft_df['Year'].astype(int)
```

```python
import pandas as pd

# Assuming your dataframe is named 'draft_df'
# Replace 'draft_df' with the actual name of your dataframe

# Create an empty dictionary to store the draft year, team, and player information
draft_dict = {}

# Iterate over the dataframe rows
for index, row in draft_df.iterrows():
    year = int(row['Year'])
    team = row['Team']
    player = row['Player']
    round = row['Round']
    pick = row['Pick']
    pos = row['Position']

    # If the draft year is not already in the outer dictionary, create a new inner dictionary
    if year not in draft_dict:
        draft_dict[year] = {}

    # If the team is not already in the inner dictionary for the draft year, create a new list for players
    if team not in draft_dict[year]:
        draft_dict[year][team] = {}

     # If the team is not already in the inner dictionary for the draft year, create a new list for players
    if 'draft' not in draft_dict[year][team]:
        draft_dict[year][team]['draft'] = []

    # Append the player to the list of players for the team and draft year
    player_dict = {"Player": player, "Round": round, "Pick": pick, "Position":pos}
    draft_dict[year][team]['draft'].append(player_dict)

# Print the draft dictionary
# for year, teams in draft_dict.items():
#     print(f"Year: {year}")
#     for team, players in teams.items():
#         print(f"Team: {team}\nPlayers: {players}\n")
```
